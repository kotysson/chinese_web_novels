```python
import os
import json
```

```python
folder_path = r'D:\Studies\HSE\project\texts_ner'
per = []
loc = []
org = []
```

```python
for filename in os.listdir(folder_path):
    file_path = os.path.join(folder_path, filename)

    with open(file_path, 'r', encoding='utf8') as file:
        data = json.load(file)

        per.extend(s.strip() for s in data.get('人名') if len(s) < 10)
        loc.extend(s.strip() for s in data.get('地名') if len(s) < 10)
        org.extend(s.strip() for s in data.get('组织') if len(s) < 10)
```

```python
for elem in per:
    elem = elem.replace('\n', '')    

for elem in loc:
    elem = elem.replace('\n', '')    

for elem in org:
    elem = elem.replace('\n', '')    
```
