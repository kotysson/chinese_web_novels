```python
import paddlehub as hub

# модель не для чистого NER, это лексический анализатор, поэтому другие теги
ner_model = hub.Module(name="lac")

with open(r'D:\Studies\HSE\project\texts\一级律师[星际].txt', "r", encoding="utf-8") as file:
  text = [file.read()]
  
# text = ["燕绥之在德卡马的黑市区寻找信息"] # текст списком
result = ner_model.lexical_analysis(texts=text)
named_entities = [(word, tag) for word, tag in zip(result[0]['word'], result[0]['tag']) if tag in {"PER", "LOC", "ORG", "ns", "nt"}]


# в рез-тах все теги, надо будет фильтровать
for word, tag in set(named_entities):
    print(f"{word}: {tag}")
```

```python
from paddlenlp import Taskflow
import pprint

# Загружаем UIE для извлечения именованных сущностей (NER)
ner = Taskflow("information_extraction", schema=["人名", "地名", "组织"])

# Анализируем текст
with open(r'D:\Studies\HSE\project\texts\一级律师[星际].txt', "r", encoding="utf-8") as file:
  text = file.read()

result = ner(text)

# Выводим результат
pprint.pprint(result)
```

```python
import os
import json

def get_named_entities(folder_path, output_folder='texts_ner'):
    ner = Taskflow("information_extraction", schema=["人名", "地名", "组织"])
    os.makedirs(output_folder, exist_ok=True)

    for filename in os.listdir(folder_path):
        file_path = os.path.join(folder_path, filename)

        with open(file_path, "r", encoding="utf-8") as file:
            text = file.read()
            result = ner(text)
            unique_texts = {}

            for entry in result:
                for key, values in entry.items():
                    if key not in unique_texts:
                        unique_texts[key] = set()
                    unique_texts[key].update(item["text"] for item in values)
            # делаем из множества список
            unique_texts = {key: list(values) for key, values in unique_texts.items()}

            output_file = os.path.join(output_folder, f"{filename}_ner.json")

            with open(output_file, "w", encoding='utf8') as outfile: 
                json.dump(unique_texts, outfile, ensure_ascii=False, indent=4)
                print(f'{output_file} created sucessfully')
```

```python
folder_path = r"D:\Studies\HSE\project\texts"
get_named_entities(folder_path)
```
